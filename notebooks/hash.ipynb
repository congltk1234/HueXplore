```python
bruh = []
import random
for i in range(len(df_mg["Tên"])):
    name = df_mg.iloc[i]["Tên"]
    # interests = [interest_mapper.get(item, item) for item in [*df_mg.iloc[i]["Interests"]]]
    # moods = [moods_mapper.get(item, item) for item in [*df_mg.iloc[i]["Mood"]]]
    img_url = df_mg.iloc[i]["Link ảnh"]
    gg_map = df_mg.iloc[i]["Googlemap"]
    coordinate = df_mg.iloc[i]["Tọa độ"].split(",")
    node_id = ox.distance.nearest_nodes(G, float(coordinate[1]), float(coordinate[0])) 
    bruh.append({
        "name": name,
        # "interests": interests,
        # "moods": moods,
        "img_url": img_url,
        "gg_map": gg_map,
        "coordinate": coordinate,
        "node_id": int(node_id)
    })
    
```
