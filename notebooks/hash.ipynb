```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
interest_mapper = {
    "1": "architecture", 
    "2": "art", 
    "3": "culture", 
    "4": "heritage", 
    "5": "history",
    "6": "market",
    "7": "museum",
    "8": "nature",
    "9": "wellness",
    "0": "ruin",
    "a": "workshop"
    }
moods_mapper = {
    "1": "roadtrip", 
    "2": "vintage", 
    "3": "romantic", 
    "4": "spiritual", 
    "5": "city explorer",
    "6": "shopping",
    "7": "photography",
    "8": "beach-loving",
    "9": "sightseeing",
    "0": "cuisine",
    "a": "relax",
    "b": "entertain",
    "c": "memorial"
}
```

```python
df_mg = df[:19]
```

```python
df_mg
```

```python
df_mg.head(1)
```

```python
bruh = []
import random
for i in range(len(df_mg["Name"])):
    name = df_mg.iloc[i]["Name"]
    interests = [interest_mapper.get(item, item) for item in [*df_mg.iloc[i]["Interests"]]]
    moods = [moods_mapper.get(item, item) for item in [*df_mg.iloc[i]["Mood"]]]
    img_url = df_mg.iloc[i]["img_url"]
    gg_map = df_mg.iloc[i]["Link_ggmap"]
    budget = random.choice(["high", "low"])
    coordinate = df_mg.iloc[i]["Tọa độ"].split(",")
    address = df_mg.iloc[i]["Địa Chỉ"]
    bruh.append({
        "name": name,
        "interests": interests,
        "moods": moods,
        "img_url": img_url,
        "gg_map": gg_map,
        "budget": budget,
        "coordinate": coordinate,
        "address": address 
    })
    
```

```python
bruh
```

```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017")
db = client["location"]
collection = db["main_loc"]


collection.insert_many(bruh)

print("heh")
```
