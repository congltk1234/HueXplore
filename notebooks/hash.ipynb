```python
interest_mapper = {
    "1": "architecture", 
    "2": "art", 
    "3": "culture", 
    "4": "heritage", 
    "5": "history",
    "6": "market",
    "7": "museum",
    "8": "nature",
    "9": "wellness",
    "0": "ruin",
    "a": "workshop"
    }
moods_mapper = {
    "1": "roadtrip", 
    "2": "vintage", 
    "3": "romantic", 
    "4": "spiritual", 
    "5": "city explorer",
    "6": "shopping",
    "7": "photography",
    "8": "beach-loving",
    "9": "sightseeing",
    "0": "cuisine",
    "a": "relax",
    "b": "entertain",
    "c": "memorial"
}
```

```python
df_loc = df
```

```python
df_loc = df_loc.sort_values(by=["img_url"])
# df_mg = df_mg[:204]
# df_mg
df_loc = df_loc[:158]
df_loc
```

```python
from flask import Flask, request, jsonify
from pymongo import MongoClient
import pandas as pd
from time import time
import pandas as pd
from pymongo import MongoClient
import osmnx as ox
G = ox.load_graphml('graph.graphml')
```

```python
df_loc.head(0)
```

```python
loc_can = []
for i in range(len(df_loc["Name"])):
    name = df_loc.iloc[i]["Name"]
    interests = [interest_mapper.get(item, item) for item in [*df_loc.iloc[i]["Interests"]]]
    moods = [moods_mapper.get(item, item) for item in [*df_loc.iloc[i]["Mood"]]]
    img_url = df_loc.iloc[i]["img_url"]
    gg_map = df_loc.iloc[i]["Link_ggmap"]
    coordinate = df_loc.iloc[i]["Tọa độ"].split(",")
    vote = df_loc.iloc[i]["Vote"]
    if name == "Chùa Tây Thiên" or name == "Warter puppet show":
        review = "0"
    else:
        review = int(df_loc.iloc[i]["Reviews"])
    price = df_loc.iloc[i]["Price"]
    duration = float(df_loc.iloc[i]["duration"])
    address = df_loc.iloc[i]["Địa Chỉ"]
    node_id = ox.distance.nearest_nodes(G, float(coordinate[1]), float(coordinate[0])) 
    loc_can.append({
        "name": name,
        "interests": interests,
        "moods": moods,
        "img_url": img_url,
        "gg_map": gg_map,
        "coordinate": coordinate,
        "vote": vote,
        "review": review,
        "price": price,
        "duration": duration,
        "address": address,
        "node_id": int(node_id)
    })
    
```

```python
loc_can
```

```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017")
db = client["unihack"]
collection = db["loc_candidate"]


collection.insert_many(loc_can)

print("heh")
```
